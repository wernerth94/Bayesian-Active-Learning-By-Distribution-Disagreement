# Sarcos

```python
fig, axes = plt.subplots(1,2, dpi=150, figsize=(12,5))
dataset = "Sarcos"
query_size = 50
kwargs = { "moving_avrg": 0.1,
           "show_auc": False,
           "plot_std": False}
axes = [axes[1], axes[0]]
# plot_batch_benchmark(axes, f"{dataset}/Oracle", "red", display_name="Oracle")
plot_batch_benchmark(axes, f"{dataset}/{query_size}/RandomAgent", "black", display_name=f"Random", **kwargs)
# plot_batch_benchmark(axes, f"{dataset}/{query_size}/NF_Conf", "green", display_name=f"NSF conf", **kwargs)
# plot_batch_benchmark(axes, f"{dataset}/{query_size}/NF_Entropy", "lime", display_name=f"NSF entropy", **kwargs)
# plot_batch_benchmark(axes, f"{dataset}/{query_size}/NF_Std", "gold", display_name=f"NSF std", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/NF_Sample_Std", "orange", display_name=f"NSF sample std", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/Coreset_Greedy", "blue", display_name=f"Coreset", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/CoreGCN", "navy", display_name=f"CoreGCN", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/TypiClust", "yellow", display_name=f"TypiClust", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALD_Point", "blueviolet", display_name=f"BALD_naive", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALD_Dist_Entropy", "green", display_name=f"BALD Dist Entropy", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALD_Dist_Std", "lime", display_name=f"BALD Dist Std", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALSA_EMD", "violet", display_name=f"BALSA_EMD", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/NFlows_Out", "red", display_name=f"NFlows Out", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALSA_KL_Diff", "khaki", display_name=f"BALSA KL Diff", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALSA_KL_Anomaly", "olive", display_name=f"BALSA Anomaly", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALSA_KL_Grid", "orange", display_name=f"BALSA KL Grid", **kwargs)
plot_batch_benchmark(axes, f"{dataset}/{query_size}/BALSA_KL_Pairs", "powderblue", display_name=f"BALSA KL Pairs", **kwargs)
plot_upper_bound(axes, dataset, 1200, "black", percentile=1.0)

axes[0].set_title('NF Sarcos MAE')
axes[1].set_title('NF Sarcos NLL')
axes[0].set_ylabel('MAE')
axes[1].set_ylabel('NLL')
axes[0].set_xlabel('# of labeled datapoints')
axes[1].set_xlabel('# of labeled datapoints')
axes[0].set_ylim(0.01, 0.03)
# axes[1].set_ylim(-1., 2.5)
plt.grid(visible=True)
axes[0].yaxis.set_major_formatter(FormatStrFormatter('%0.3f'))
axes[0].legend(fontsize='x-small', loc='upper left', bbox_to_anchor=(-1, 1))
fig.tight_layout()
fig.subplots_adjust(right=1.0, hspace=0.2, wspace=1.)
plt.savefig(f"paper/img/traj_nf_sarcos_{query_size}.pdf")
plt.show()
```
